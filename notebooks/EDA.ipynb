```python
from haversine import haversine

# Times Square coordinates
times_square = (40.7580, -73.9855)

def get_distance(row):
    user_location = (row['latitude'], row['longitude'])
    return haversine(user_location, times_square)

df['distance_to_times_square'] = df.apply(get_distance, axis=1)
```

```python
plt.figure(figsize=(8, 4))
sns.scatterplot(x='distance_to_times_square', y='price', data=df)
plt.title("Price vs Distance to Times Square")
plt.ylim(0, 500)
plt.xlabel("Distance to Times Square (km)")
plt.show()
```

```python
print(df[['price', 'distance_to_times_square']].corr())
```

```python
# Boxplot: Price by Room Type
plt.figure(figsize=(8, 4))
sns.boxplot(x='room_type', y='price', data=df)
plt.title("Price Distribution by Room Type")
plt.ylim(0, 500)
plt.show()
```

```python
df_clean = df.copy()  # make a copy for comparison
df_clean.head()
```

```python
print("Missing values before cleaning:\n", df_clean.isnull().sum())
```

```python
# Fill missing last_review with 'No Review'
df_clean['last_review'] = df_clean['last_review'].fillna('No Review')

# Fill missing reviews_per_month with 0
df_clean['reviews_per_month'] = df_clean['reviews_per_month'].fillna(0)
```

```python
df_clean['last_review'] = pd.to_datetime(df_clean['last_review'], errors='coerce')
```

```python
duplicates = df_clean.duplicated().sum()
df_clean.drop_duplicates(inplace=True)
print(f"Removed {duplicates} duplicate rows.")
```

```python
before_price_zero = (df_clean['price'] == 0).sum()
df_clean = df_clean[df_clean['price'] > 0]
print(f"Removed {before_price_zero} listings with price = 0.")
```

```python
before_min_nights = (df_clean['minimum_nights'] > 30).sum()
df_clean = df_clean[df_clean['minimum_nights'] <= 30]
print(f"Removed {before_min_nights} listings with minimum_nights > 30.")
```

```python
df_clean = df_clean[df_clean['price'] <= 500]
df_clean = df_clean[df_clean['number_of_reviews'] <= 200]
df_clean = df_clean[df_clean['reviews_per_month'] <= 10]
```

```python
df_clean['days_since_last_review'] = (
    pd.to_datetime('2019-07-10') - df_clean['last_review']
).dt.days.fillna(999)
```

```python
print("Missing values after cleaning:\n", df_clean.isnull().sum())
```

```python
print("Shape before:", df.shape)
print("Shape after :", df_clean.shape)

print("\nSample before cleaning:")
print(df.head())

print("\nSample after cleaning:")
print(df_clean.head())
```
